```python
#HOME TASK 1;
#Home Tasks:
#Question #1(Fuzzification of Beverage Temperature)
#A smart coffee machine categorizes the temperature of a beverage into three fuzzy sets: "Too Cold," "Optimal," and "Too Hot." For a beverage at 50°C, compute the degree of membership for each set using triangular membership functions. Visualize the membership values for temperatures ranging from 0°C to 100°C
#
!pip install scikit-fuzzy
import numpy as np
import matplotlib.pyplot as plt

# Temperature range
T = np.linspace(0, 100, 1000)

# Membership functions
too_cold = np.maximum(0, (40 - T) / 40)
optimal = np.maximum(0, np.minimum((T - 30) / 20, (70 - T) / 20))
too_hot = np.maximum(0, (T - 60) / 40)

# Plot
plt.figure()
plt.plot(T, too_cold, label="Too Cold")
plt.plot(T, optimal, label="Optimal")
plt.plot(T, too_hot, label="Too Hot")

plt.xlabel("Temperature (°C)")
plt.ylabel("Membership Degree")
plt.title("Fuzzification of Beverage Temperature")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#HOME TASK 2
#Question #2(Defuzzification for Car Speed Control)
#A self-driving car adjusts its speed based on road conditions using fuzzy rules: If "Slippery," reduce speed proportionally.
#If "Normal," maintain speed.
#If "Dry," increase speed proportionally.
#For a road condition with a slipperiness value of 0.6, compute the crisp output using the weighted average defuzzification method.

import matplotlib.pyplot as plt

# Road conditions
conditions = ['Slippery', 'Normal', 'Dry']

# Membership values
membership = [0.6, 0.3, 0.1]

# Corresponding speed values (km/h)
speed = [40, 60, 80]

# Weighted contribution
weighted_speed = [membership[i] * speed[i] for i in range(3)]

# Crisp output (Weighted Average)
crisp_speed = sum(weighted_speed) / sum(membership)

# Plot bar graph
plt.figure()
plt.bar(conditions, speed)
plt.axhline(y=crisp_speed, linestyle='--', label='Crisp Speed Output')

plt.xlabel('Road Condition')
plt.ylabel('Speed (km/h)')
plt.title('Defuzzification using Weighted Average Method')
plt.legend()
plt.grid(True)

plt.show()
```
