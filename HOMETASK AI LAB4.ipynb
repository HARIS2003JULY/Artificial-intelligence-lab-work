```python
#LAB 4 
#HOME TASK
#Question #1: Chocolate Factory – The Path to the Secret Recipe

#Willy Wonka’s Chocolate Factory is filled with rooms connected through chocolate tunnels. A worker robot must explore the factory to reach the Secret Recipe Vault where the ultimate chocolate formula is stored.However, some tunnels are blocked due to maintenance or melted chocolate spills, making certain routes inaccessible.

#The robot uses Depth-First Search (DFS) to explore all possible paths from the Entrance to the Secret Vault, avoiding blocked routes.

#•	Draw the state-space representation (graph of rooms and tunnels) showing blocked tunnels.
#•	Implement Depth-First Search (DFS) to explore all possible routes from Entrance → Secret Vault.
#•	Display the visited order of rooms and all valid DFS paths avoiding blocked tunnels.

# Depth-First Search for Chocolate Factory

# Graph representation (adjacency list)
graph = {
    'A': ['B', 'D'],
    'B': ['A', 'C'],
    'C': ['B', 'G'],
    'D': ['A', 'E'],
    'E': ['D', 'F'],
    'F': ['E'],
    'G': ['C']
}

visited = []  # to record visited rooms
all_paths = []  # to store all valid paths

def dfs(current, target, path):
    visited.append(current)
    path.append(current)

    if current == target:
        all_paths.append(list(path))
    else:
        for neighbor in graph[current]:
            if neighbor not in path:  # avoid revisiting rooms
                dfs(neighbor, target, path)

    path.pop()

# Run DFS from Entrance (A) to Secret Vault (G)
dfs('A', 'G', [])

print("Visited Order:", visited)
print("All Valid DFS Paths (Entrance → Secret Vault):")
for p in all_paths:
    print(" → ".join(p))

```
