```python
#lab 4
                       #class task
#task 1
#Class tasks:
#Task 1: Robot Maze Search using DFS
#A   firefighting   robot   is   trapped   in   a   maze-like   4×4   building   floor. It  must  find  the  fire  location  (3,  3)  starting  from  entrance  (0,  0). Blocked    rooms    represent    debris,    which    cannot    be    crossed. The   robot   moves   Up,   Down,   Left,   Right   within   the   grid. Use DFS to explore all possible routes to the fire.z
#Implement all these in your tasks

#1.	Draw the state-space / adjacency list for all reachable rooms.
#2.	Implement DFS to explore all rooms starting from the entrance.
#3.	Find and display all possible paths from entrance to fire location.
# Task 1 – DFS Robot Maze Search 
edges = { 
    (0,0): [(0,1),(1,0)], 
    (0,1): [(0,0),(0,2)], 
    (0,2): [(0,1),(0,3),(1,2)], 
    (0,3): [(0,2),(1,3)], 
    (1,0): [(0,0),(2,0)], 
    (1,2): [(0,2),(1,3)], 
    (1,3): [(0,3),(1,2),(2,3)], 
    (2,0): [(1,0),(3,0)], 
    (2,3): [(1,3),(3,3)], 
    (3,0): [(2,0),(3,1)], 
    (3,1): [(3,0),(3,2)], 
    (3,2): [(3,1),(3,3)], 
    (3,3): [(3,2),(2,3)] 
} 
 
print("Adjacency List (State-Space Representation):") 
for k,v in edges.items(): 
    print(f"{k} → {v}") 
 
def dfs_all_paths(graph, start, goal): 
    stack = [(start, [start])] 
    visited = [] 
    all_paths = [] 
    while stack: 
        node, path = stack.pop() 
        if node not in visited: 
            visited.append(node) 
        if node == goal: 
            all_paths.append(path) 
        for neighbor in graph.get(node, []): 
            if neighbor not in path: 
                stack.append((neighbor, path + [neighbor])) 
    return visited, all_paths 
 
visited, paths = dfs_all_paths(edges, (0,0), (3,3))
print("\nDFS Exploration Order:") 
print(visited) 
print("\nAll Possible Paths from Entrance to Fire using DFS:") 
for i,p in enumerate(paths,1): 
    print(f"Path {i}: {p}")
```

```python
#TASK 2
#Task 2: City Network Exploration using DFS 
#Department of Computer Science      
#• A logistics company needs to analyze all possible road routes connecting major cities. 
#Each city is a node, and roads are edges. 
#Use DFS to explore all routes from Karachi to Faisalabad.
#Implement all these in your tasks: 
#1. Draw the state-space / adjacency list for all reachable rooms. 
#2. Implement DFS to explore all rooms starting from the entrance. 
#3. Find and display all possible paths from entrance to fire location.


city_graph = { 
    "Karachi": ["Lahore","Islamabad"], 
    "Lahore": ["Karachi","Peshawar","Faisalabad"], 
    "Islamabad": ["Karachi"], 
    "Peshawar": ["Lahore"], 
    "Faisalabad": ["Lahore"] 
} 
 
print("Adjacency List (City Network):") 
for k,v in city_graph.items(): 
    print(f"{k} → {v}") 
 
def dfs_all_paths_city(graph, start, goal): 
    stack = [(start,[start])] 
    visited = [] 
    all_paths = [] 
    while stack: 
        node,path = stack.pop() 
        if node not in visited: 
            visited.append(node) 
        if node == goal: 
            all_paths.append(path) 
        for nbr in graph[node]: 
            if nbr not in path: 
                stack.append((nbr,path+[nbr])) 
    return visited,all_paths 
 
visited, paths = dfs_all_paths_city(city_graph,"Karachi","Faisalabad") 
 
print("\nDFS Exploration Order (Cities Visited):") 
print(visited) 
print("\nAll Possible Paths from Karachi to Faisalabad:") 
for i,p in enumerate(paths,1): 
    print(f"Path {i}: {p}")
```
