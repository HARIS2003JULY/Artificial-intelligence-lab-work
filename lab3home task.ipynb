```python
#lab 3 home task
#home task 1
#Question #1: Using BFS to Find the Shortest Path
#i. In the Chocolate Factory, the robot needs to find the shortest path to deliver raw materials from
#the storage area (position S) to the production line (position G). The grid layout includes
#obstacles (X) blocking some paths, and the robot can only move up, down, left, or right.
#ii. How would Breadth-First Search (BFS) find the shortest path from S to G in this chocolate
#factory grid,
#iii. Hint: BFS works level-by-level, guaranteeing the shortest path in unweighted grids and
#avoiding deep exploration of unnecessary paths.

from collections import deque

def bfs_chocolate_factory(grid, start, goal):
    rows, cols = len(grid), len(grid[0])
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]  # Right, Down, Left, Up
    queue = deque([(start, [start])])  # (current_position, path_so_far)
    visited = set()

    while queue:
        (x, y), path = queue.popleft()

        # If goal reached, return the path
        if (x, y) == goal:
            return path

        if (x, y) not in visited:
            visited.add((x, y))

            for dx, dy in directions:
                nx, ny = x + dx, y + dy
                if 0 <= nx < rows and 0 <= ny < cols and grid[nx][ny] != "X" and (nx, ny) not in visited:
                    queue.append(((nx, ny), path + [(nx, ny)]))
    
    return None  # No path found


# ðŸ­ Chocolate Factory Layout (Grid)
factory_grid = [
    ["S", ".", ".", "X", "G"],
    [".", "X", ".", ".", "."],
    [".", ".", ".", "X", "."],
    ["X", ".", ".", ".", "."],
    [".", ".", "X", ".", "."]
]

start = (0, 0)  # Position of S (Storage)
goal = (0, 4)   # Position of G (Production Line)

# Run BFS
shortest_path = bfs_chocolate_factory(factory_grid, start, goal)

print("Chocolate Factory Layout:")
for row in factory_grid:
    print(" ".join(row))

print("\nShortest Path from S to G:")
print(shortest_path if shortest_path else "No path found")
```

```python
#home task 2
#Question #2(Analyze BFS on a Grid with Varying Obstacles)
#i. In Karachi, a delivery person must navigate between two restaurants in a busy area. The
#area is mapped as a grid, where each cell represents either a road or an obstacle (e.g.,
#buildings, traffic). The goal is to find the shortest path between the two restaurants while
#avoiding obstacles. The grid changes with different obstacle densities, ranging from low to
#medium and high.
#ii. Test the performance of Breadth-First Search (BFS) on grids with varying densities of
#obstacles (low, medium, high).
#iii. Measure and compare the execution time and path length for each grid.

import time
import random
from collections import deque

# BFS Algorithm
def bfs_navigation(grid, start, goal):
    rows, cols = len(grid), len(grid[0])
    directions = [(0,1), (1,0), (0,-1), (-1,0)]  # Right, Down, Left, Up
    queue = deque([(start, [start])])
    visited = set()

    while queue:
        (x, y), path = queue.popleft()

        if (x, y) == goal:
            return path  # Found shortest path

        if (x, y) not in visited:
            visited.add((x, y))

            for dx, dy in directions:
                nx, ny = x + dx, y + dy
                if 0 <= nx < rows and 0 <= ny < cols and grid[nx][ny] != "X" and (nx, ny) not in visited:
                    queue.append(((nx, ny), path + [(nx, ny)]))

    return None  # No path found


# Generate Grid with Random Obstacles
def generate_grid(size, obstacle_density):
    grid = [["." for _ in range(size)] for _ in range(size)]
    for i in range(size):
        for j in range(size):
            if random.random() < obstacle_density:
                grid[i][j] = "X"
    grid[0][0] = "S"
    grid[size-1][size-1] = "G"
    return grid


# Function to test BFS on different obstacle densities
def test_bfs_on_obstacles():
    start = (0, 0)
    goal = (9, 9)
    densities = {"Low (20%)": 0.2, "Medium (40%)": 0.4, "High (60%)": 0.6}

    for level, density in densities.items():
        grid = generate_grid(10, density)
        start_time = time.time()
        path = bfs_navigation(grid, start, goal)
        end_time = time.time()
        exec_time = end_time - start_time
        path_length = len(path) if path else 0

        print(f"\n--- {level} Obstacle Grid ---")
        for row in grid:
            print(" ".join(row))
        print(f"Execution Time: {exec_time:.6f} seconds")
        print(f"Path Length: {path_length if path_length else 'No path found'}")


# Run Tests
test_bfs_on_obstacles()
```

```python
#home task 3
#Question #3(Apply BFS for COCONUT GROVE)
#In the heart of Karachi, a delivery person needs to measure exactly 4 liters of water using two
#travel flasks to unlock a special dish at coconut grove. Flask 1 has a capacity of 3 liters and
#Flask 2 has a capacity of 6 liters. The task is to measure the exact amount of 4 liters of water.
#i. Implement Breadth-First Search (BFS) to solve the water jug problem. The target is to
#measure exactly 3 liters instead of 6 liters as in the initial problem.
#ii. Develop an algorithm that finds the shortest sequence of steps to measure 4 liters using BFS

from collections import deque

def bfs_coconut_grove(cap1, cap2, target):
    visited = set()          # Track visited states
    parent = {}              # To trace the path
    queue = deque([(0, 0)])  # Start with both flasks empty

    while queue:
        f1, f2 = queue.popleft()

        # Check if target reached
        if f1 == target or f2 == target:
            path = []
            while (f1, f2) in parent:
                path.append((f1, f2))
                f1, f2 = parent[(f1, f2)]
            path.append((0, 0))
            return path[::-1]  # Reverse to get start â†’ goal

        if (f1, f2) not in visited:
            visited.add((f1, f2))

            # All possible actions
            moves = [
                (cap1, f2),  # Fill Flask 1
                (f1, cap2),  # Fill Flask 2
                (0, f2),     # Empty Flask 1
                (f1, 0),     # Empty Flask 2
                # Pour Flask 1 â†’ Flask 2
                (f1 - min(f1, cap2 - f2), f2 + min(f1, cap2 - f2)),
                # Pour Flask 2 â†’ Flask 1
                (f1 + min(f2, cap1 - f1), f2 - min(f2, cap1 - f1))
            ]

            # Explore each new move
            for move in moves:
                if move not in visited:
                    queue.append(move)
                    parent[move] = (f1, f2)

    return None  # No solution found


# ðŸ¥¥ COCONUT GROVE Problem Setup
capacity1 = 3
capacity2 = 6
target = 4

print("BFS Solution Path for Coconut Grove (Water Jug Problem):")
solution_path = bfs_coconut_grove(capacity1, capacity2, target)
print(solution_path if solution_path else "No solution found")
```
