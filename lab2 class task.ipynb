```python
#lab 2
#class task 
#task 1
#Task 1: (smoke detection system based on simple reflex agent):              
#Design and implementation of a simple reflex agent that detects smoke and triggers 
#an alert. The agent will operate based on a set of condition-action rules, responding 
#to the presence of smoke in the environment. 

import time
import random

class SmokeDetector:
    def __init__(self):
        self.alert_active = False

    def read_sensor(self):
        # Simulate smoke detection (10% chance of detecting smoke)
        return random.choice([True] * 1 + [False] * 9)

    def activate_alert(self):
        self.alert_active = True
        print("Alert! Smoke Detected!")

    def deactivate_alert(self):
        self.alert_active = False
        print("No smoke detected. All clear.")

    def run(self):
        print("Smoke Detector System is active.")
        while True:
            smoke_detected = self.read_sensor()
            if smoke_detected:
                self.activate_alert()
            else:
                self.deactivate_alert()
            time.sleep(0)  # Check every second


# Run the smoke detector
if __name__ == "__main__":
    detector = SmokeDetector()
    detector.run()
```

```python
#task2
#Task 2: :(project management system based on goal based agent): 
#To create a goal-based agent that manages tasks in a project 
#management system, ensuring that tasks are completed in a prioritized manner to 
#meet project deadlines. 

class Task:
    def __init__(self, name, priority, deadline):
        self.name = name
        self.priority = priority
        self.deadline = deadline
        self.completed = False


class ProjectManagementAgent:
    def __init__(self):
        self.tasks = []

    def add_task(self, task):
        self.tasks.append(task)

    def choose_next_task(self):
        # Sort tasks by priority (higher priority first) and then by deadline
        self.tasks.sort(key=lambda x: (-x.priority, x.deadline))
        for task in self.tasks:
            if not task.completed:
                return task
        return None

    def complete_task(self, task):
        task.completed = True
        print(f"Completed task: {task.name}")

    def run(self):
        while True:
            task = self.choose_next_task()
            if task is None:
                print("All tasks are completed!")
                break
            self.complete_task(task)


# --- Set up the project management agent ---
if __name__ == "__main__":
    agent = ProjectManagementAgent()

    # Adding tasks with (name, priority, deadline)
    agent.add_task(Task("Design goal based agent", priority=2, deadline="2024-10-20"))
    agent.add_task(Task("Develop model for goal based agent", priority=1, deadline="2024-10-25"))
    agent.add_task(Task("Create project management system", priority=3, deadline="2024-10-30"))

    agent.run()
```

```python
#task 3
#Task 3: :(content recommendation system based on learning-based agent): 
#Create an interactive simulation of a learning-based recommendation system that 
#adapts user preferences over time, similar to platforms like Netflix and Amazon. 
#Participants will understand how learning agents adjust recommendations based on 
#changing user behaviors.. 

import random
import time

class Content:
    def __init__(self, title, genre):
        self.title = title
        self.genre = genre

class User:
    def __init__(self, name):
        self.name = name
        # User preference weights for each genre
        self.preferences = {
            "Action": 0.5,
            "Comedy": 0.5,
            "Drama": 0.5,
            "Horror": 0.5,
            "Romance": 0.5
        }

    def watch_content(self, content):
        """Simulate user feedback (like/dislike) based on preferences."""
        # Probability of liking depends on user's preference for that genre
        like_probability = self.preferences[content.genre]
        liked = random.random() < like_probability
        return liked

    def update_preferences(self, content, liked):
        """Learning step: Adjust preferences based on feedback."""
        if liked:
            # Strengthen preference for liked genre
            self.preferences[content.genre] = min(1.0, self.preferences[content.genre] + 0.1)
        else:
            # Decrease preference for disliked genre
            self.preferences[content.genre] = max(0.0, self.preferences[content.genre] - 0.1)

class LearningRecommendationAgent:
    def __init__(self, user, contents):
        self.user = user
        self.contents = contents

    def recommend(self):
        """Recommend content based on user's current preferences."""
        # Rank contents by user's preference for the genre
        sorted_contents = sorted(self.contents, key=lambda c: self.user.preferences[c.genre], reverse=True)
        return sorted_contents[0]  # Recommend the top one

    def run(self, rounds=10):
        print(f"üé¨ Welcome {self.user.name}! Starting personalized content recommendations...\n")
        for i in range(rounds):
            content = self.recommend()
            print(f"‚û°Ô∏è  Recommended: {content.title} ({content.genre})")

            liked = self.user.watch_content(content)
            if liked:
                print(f"‚úÖ You liked '{content.title}'!")
            else:
                print(f"‚ùå You didn‚Äôt enjoy '{content.title}'.")

            # Learning update
            self.user.update_preferences(content, liked)
            print(f"Updated preferences: {self.user.preferences}\n")

            # Random delay for realism
            time.sleep(1)

        print("üéâ Simulation complete! Final learned preferences:")
        for genre, weight in self.user.preferences.items():
            print(f"   {genre}: {round(weight, 2)}")


# --- Simulation Setup ---
if __name__ == "__main__":
    # Sample content library
    contents = [
        Content("Fast & Furious", "Action"),
        Content("The Hangover", "Comedy"),
        Content("Titanic", "Romance"),
        Content("The Conjuring", "Horror"),
        Content("Breaking Bad", "Drama"),
        Content("Avengers", "Action"),
        Content("Friends", "Comedy"),
    ]

    # Create user and agent
    user = User("Haris")
    agent = LearningRecommendationAgent(user, contents)

    # Run the learning-based recommendation simulation
    agent.run(rounds=7)
```
