```python
#lab 1
#home task
#task 1
#Home task #1 
#Create a class Student with private attributes for name and grade. Provide methods to get and 
#set these values securely. 

class Student:
    # Constructor to initialize private attributes
    def __init__(self, name, grade):
        self.__name = name      # private attribute
        self.__grade = grade    # private attribute

    # Getter for name
    def get_name(self):
        return self.__name

    # Setter for name
    def set_name(self, name):
        if isinstance(name, str) and name.strip() != "":
            self.__name = name
        else:
            print("Invalid name! Please enter a valid string.")

    # Getter for grade
    def get_grade(self):
        return self.__grade

    # Setter for grade
    def set_grade(self, grade):
        if 0 <= grade <= 100:
            self.__grade = grade
        else:
            print("Invalid grade! Must be between 0 and 100.")


# --- Testing the class ---
student1 = Student("Ali", 85)

print("Name:", student1.get_name())
print("Grade:", student1.get_grade())

student1.set_name("Haris")
student1.set_grade(92)

print("Updated Name:", student1.get_name())
print("Updated Grade:", student1.get_grade())
```

```python
#task 2
#Home task #2 
#Implement a base class Vehicle and derive Car and Bike classes. Demonstrate polymorphism 
#using a method vehicle_type.. 

# Base class
class Vehicle:
    def vehicle_type(self):
        print("This is a generic vehicle.")

# Derived class Car
class Car(Vehicle):
    def vehicle_type(self):
        print("This is a Car.")

# Derived class Bike
class Bike(Vehicle):
    def vehicle_type(self):
        print("This is a Bike.")

# --- Demonstrating Polymorphism ---
# Same method name (vehicle_type) behaves differently depending on the object type

vehicles = [Car(), Bike(), Vehicle()]

for v in vehicles:
    v.vehicle_type()
```

```python
#task 3
#Home task #3 
#Create a bank account system using encapsulation to manage account balance securely. 
class BankAccount:
    def __init__(self, account_holder, balance=0):
        self.__account_holder = account_holder   # Private attribute
        self.__balance = balance                 # Private attribute

    # Getter for account holder
    def get_account_holder(self):
        return self.__account_holder

    # Getter for balance
    def get_balance(self):
        return self.__balance

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive!")

    # Method to withdraw money
    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be positive!")
        elif amount > self.__balance:
            print("Insufficient funds!")
        else:
            self.__balance -= amount
            print(f"Withdrew ${amount}. Remaining balance: ${self.__balance}")

# --- Testing the BankAccount class ---

account1 = BankAccount("Haris", 1000)

print("Account Holder:", account1.get_account_holder())
print("Initial Balance:", account1.get_balance())

account1.deposit(500)
account1.withdraw(200)
account1.withdraw(2000)  # Trying to withdraw more than available balance
```
