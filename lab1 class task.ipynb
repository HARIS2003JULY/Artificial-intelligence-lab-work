```python

# lab 1 
# class task 
# task 1
#Task 1:Understanding Classes and Objects 
#Design a class to represent parrots with attributes for name and age, and 
#demonstrate object creation and attribute access. 


class Parrot: 
    # Class attributes 
    name = "" 
    age = 0 
 
# Create instances 
parrot1 = Parrot() 
parrot1.name = "Blu" 
parrot1.age = 10 
parrot2 = Parrot() 
parrot2.name = "Woo" 
parrot2.age = 15 
# Access attributes 
print(f"{parrot1.name} is {parrot1.age} years old") 
print(f"{parrot2.name} is {parrot2.age} years old") 
```

```python
#task 2
# Task 2: Implementing Inheritance: 
#Create a base class Animal with methods for eat and sleep, and a 
#derived class Dog with an additional method bark.. 
# Base class 
class Animal: 
    def eat(self): 
        print("I can eat!") 
    def sleep(self): 
        print("I can sleep!") 
# Derived class 
class Dog(Animal): 
    def bark(self): 
        print("I can bark! Woof woof!!") 
# Object of derived class 
dog1 = Dog() 
dog1.eat() 
dog1.sleep() 
dog1.bark()      
```

```python
#task 3
#Task 3: Encapsulation: 
#Implement private attributes in a Computer class and modify them through setter 
#methods.
class Computer: 
    def __init__(self): 
        self.__maxprice = 900 
    def sell(self): 
        print(f"Selling Price: {self.__maxprice}") 
    def setMaxPrice(self, price): 
        self.__maxprice = price 
# Example usage 
c = Computer() 
c.sell() 
# Attempting to change price directly 
c.__maxprice = 1000 
c.sell() 
# Changing price via setter 
c.setMaxPrice(1000) 
c.sell() 
```

```python
#task 4
#Task 4: Polymorphism: 
#Demonstrate method overriding by creating a base class Polygon and derived 
#classes Square and Circle. 
class Polygon: 
    def render(self): 
        print("Rendering Polygon...") 
 
class Square(Polygon): 
    def render(self): 
        print("Rendering Square...") 
 
class Circle(Polygon): 
    def render(self): 
        print("Rendering Circle...") 
 
# Objects of derived classes 
s1 = Square() 
s1.render() 
 
c1 = Circle() 
c1.render() 
```
