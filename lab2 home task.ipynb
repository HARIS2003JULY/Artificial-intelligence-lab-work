```python
#lab 2 
#home task 
#task 1
#Home task #1 
#Simulate a simple reflex agent that automatically opens a door when it detects a person 
#approaching. 

import time
import random

class DoorAgent:
    def __init__(self):
        self.door_open = False

    def detect_person(self):
        """Simulate person detection (30% chance someone is near)."""
        return random.choice([True] * 3 + [False] * 7)

    def open_door(self):
        if not self.door_open:
            self.door_open = True
            print("ðŸšª Door opened. Person detected!")

    def close_door(self):
        if self.door_open:
            self.door_open = False
            print("ðŸšª Door closed. No one nearby.")

    def run(self):
        print("ðŸ¤– Simple Reflex Agent started (Automatic Door System)\n")
        for i in range(10):  # Simulate 10 checks
            person_near = self.detect_person()

            # Reflex action: instant response to environment condition
            if person_near:
                self.open_door()
            else:
                self.close_door()

            time.sleep(1)

        print("\nSimulation complete âœ…")


# --- Run the Reflex Agent ---
if __name__ == "__main__":
    agent = DoorAgent()
    agent.run()
```

```python
#task2
#Home task#2 
#Simulate a goal-based AI character that strategizes to achieve specific objectives within a video 
#game, such as defending a location or defeating an opponent. 

import random
import time

class GameEnvironment:
    """Simulates a simple game world with an enemy and a base."""
    def __init__(self):
        self.enemy_distance = random.randint(1, 10)  # 1 = close, 10 = far
        self.base_health = 100

    def update(self):
        """Randomly change environment state to simulate dynamics."""
        self.enemy_distance = random.randint(1, 10)
        # Random chance of the base being attacked
        if random.random() < 0.3:
            damage = random.randint(5, 15)
            self.base_health -= damage
            print(f"âš ï¸  Base under attack! Lost {damage} HP (Health: {self.base_health})")
        else:
            print("âœ… Base is safe right now.")

class GoalBasedAgent:
    """A goal-based agent that acts strategically based on conditions."""
    def __init__(self, goal):
        self.goal = goal

    def perceive(self, environment):
        """Observe the current environment."""
        return {
            "enemy_distance": environment.enemy_distance,
            "base_health": environment.base_health
        }

    def decide_action(self, perception):
        """Decide what to do based on the goal and environment state."""
        if self.goal == "defend base":
            if perception["base_health"] < 50:
                return "repair base"
            elif perception["enemy_distance"] < 4:
                return "attack enemy near base"
            else:
                return "patrol area"

        elif self.goal == "defeat enemy":
            if perception["enemy_distance"] <= 3:
                return "attack enemy"
            else:
                return "move closer to enemy"

        else:
            return "idle"

    def act(self, action):
        """Perform the chosen action."""
        print(f"ðŸŽ¯ Action taken: {action}\n")

    def run(self, environment, rounds=10):
        """Simulate the agent's decision-making over multiple turns."""
        print(f"ðŸ¤– Goal-Based Agent initialized with goal: '{self.goal}'\n")
        for step in range(rounds):
            print(f"--- Turn {step + 1} ---")
            perception = self.perceive(environment)
            print(f"Perceived â†’ Enemy Distance: {perception['enemy_distance']}, Base Health: {perception['base_health']}")
            
            action = self.decide_action(perception)
            self.act(action)

            environment.update()
            print()
            time.sleep(1)

        print("ðŸ Simulation complete!")
```

```python
#task 3
#Home Task #3 
#â€¢ To simulate a learning-based agent that improves its understanding of user speech over time, 
#adapting to accents and slang.

import random
import time

class SpeechAgent:
    def __init__(self):
        # Known words and their meanings
        self.vocabulary = {
            "hello": "greeting",
            "bye": "farewell",
            "thanks": "gratitude",
            "music": "play_music",
            "weather": "get_weather"
        }

        # Memory of misunderstood words that can be learned later
        self.learning_memory = {}

    def hear(self):
        """Simulate user speech input (some known, some new/slang words)."""
        possible_inputs = [
            "hello", "bye", "thanks", "music", "weather",
            "yo", "sup", "chao", "thx", "meteo", "hiya", "gudbye"
        ]
        return random.choice(possible_inputs)

    def understand(self, word):
        """Try to understand the speech input."""
        if word in self.vocabulary:
            meaning = self.vocabulary[word]
            print(f"ðŸ§  Understood '{word}' as '{meaning}'.")
            return True
        else:
            print(f"â“ Did not understand '{word}'. Asking user for meaning...")
            self.learn(word)
            return False

    def learn(self, word):
        """Simulate learning from user feedback."""
        # Simulate user explaining meaning
        possible_meanings = {
            "yo": "greeting",
            "sup": "greeting",
            "chao": "farewell",
            "thx": "gratitude",
            "meteo": "get_weather",
            "hiya": "greeting",
            "gudbye": "farewell"
        }

        meaning = possible_meanings.get(word, "unknown")
        if meaning != "unknown":
            print(f"âœ… Learned that '{word}' means '{meaning}'.")
            self.vocabulary[word] = meaning
        else:
            print(f"âš ï¸ Could not learn meaning for '{word}' (unknown slang).")

    def run(self, rounds=10):
        """Simulate multiple speech interactions with user."""
        print("ðŸŽ¤ Learning-Based Speech Agent Started!\n")
        for i in range(rounds):
            word = self.hear()
            print(f"ðŸ‘‚ Heard: '{word}'")
            understood = self.understand(word)
            if understood:
                print(f"ðŸ¤– Responding appropriately to '{word}'.\n")
            else:
                print(f"ðŸ¤– Updated vocabulary after learning: {list(self.vocabulary.keys())}\n")

            time.sleep(1)

        print("ðŸ§© Final learned vocabulary:")
        for w, m in self.vocabulary.items():
            print(f"  '{w}' â†’ {m}")
```
