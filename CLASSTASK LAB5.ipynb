```python
#lab 5
#                                                     class task
#Task 1: A* Searching on Academic Program

#You are tasked with finding the most efficient route between different academic programs at Iqra University using the A* algorithm.Each program is a node in a graph, and the edges between them represent the cost of transitioning between programs.The goal is to find the most efficient (shortest) path from BS(CS) to BS(IT).
#Implement all these in your tasks
#1.	Calculate heuristic and cost.
#2.	Implement the A* algorithm to find the shortest path from BS(CS) to BS(IT).
#3.	Display the results
#a)	Shortest path
#b)	Visited nodes in order

import heapq 
 
# Graph and heuristics 
graph = { 
    'BS(CS)': {'BS(SE)': 2, 'BS(CB)': 4}, 
    'BS(SE)': {'BS(CS)': 2, 'BS(AI)': 1, 'BS(IT)': 3}, 
    'BS(AI)': {'BS(SE)': 1, 'BS(CB)': 6}, 
    'BS(CB)': {'BS(CS)': 4, 'BS(SE)': 6, 'BS(IT)': 7}, 
    'BS(IT)': {'BS(SE)': 3, 'BS(CB)': 7} 
} 
heuristic = {'BS(CS)': 9, 'BS(SE)': 8, 'BS(AI)': 6, 'BS(CB)': 4, 
'BS(IT)': 10} 
def a_star(start, goal): 
    open_list = [(heuristic[start], 0, start, [start])] 
    visited = set() 
 
    while open_list: 
        f, g, current, path = heapq.heappop(open_list) 
        visited.add(current) 
        if current == goal: 
            return path, visited 
        for neighbor, cost in graph[current].items(): 
            if neighbor not in visited: 
                g_new = g + cost 
                f_new = g_new + heuristic[neighbor] 
                heapq.heappush(open_list, (f_new, g_new, neighbor, path + [neighbor])) 
    return None, visited 
path, visited = a_star('BS(CS)', 'BS(IT)') 
print("Shortest Path:", path) 
print("Visited Nodes:", visited) 
```
